Importing Matplotlib, Seaborn, and Pandas

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame
import pandas as pd
```

Creating and Displaying a Sample DataFrame with Names, Age, Gender, and Rank

```python
data = {
    'names': ['steve', 'john', 'richard', 'sarah', 'randy', 'micheal', 'julie', 'mark'],
    'age': [20, 22, 20, 21, 24, 23, 22, 50],
    'gender': ['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male'],
    'rank': [2, 1, 4, 5, 3, 7, 6, 8]
}
df = DataFrame(data)
df
```

Bar Chart of Names vs. Age

```python
plt.bar(df['names'],df['age'])
```

```python
plt.bar(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title("Comparing names and age")
plt.show()
```

Bar Chart: Comparing Names and Their Ranks

```python
plt.bar(df['names'], df['rank'])
plt.xlabel('Names')
plt.ylabel('Rank')
plt.title("Comparing name and rank")
plt.show()
```

Bar Plot: Comparison of Names and Ages

```python
sns_bar_graph = sns.barplot(data = df, x = "names", y = "age")
sns_bar_graph.set_title("Comparing names and age")
plt.show()
```

Line Plot: Comparison of Names and Ages

```python
plt.plot(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title("Comparing names and age")
plt.show()
```

Line Plot: Comparison of Names and Ages Using Seaborn

```python
sns_line_plot = sns.lineplot(data= df, x = "names", y = "age")
sns_line_plot.set_title("Comparing names and age")
plt.show()
```

Pie Chart: Age Comparison by Names

```python
plt.pie(df['age'], labels= df["names"])
plt.title("Age comparision")
plt.show()
```

Tips Dataset: Restaurant Tipping Data Analysis

```python
tips_data = sns.load_dataset("tips")
tips_data
```

Bar Plot: Total Bill Comparison by Gender

```python
sns.barplot(data = tips_data, x = 'sex', y = 'total_bill')
```

Bar Plot: Total Bill Comparison by Gender and Smoking Status

```python
sns.barplot(data= tips_data , x = 'sex', y = 'total_bill', hue= 'smoker')
```

Count Plot: Smoking Status by Gender

```python
sns.countplot(x = tips_data['smoker'], hue= tips_data['sex'])
```

Scatter Plot: Relationship Between Total Bill and Tip Amount

```python
sns.scatterplot(data= tips_data, x = 'total_bill', y = 'tip')
```

Regression Plot: Correlation Between Total Bill and Tip Amount

```python
sns.regplot(data= tips_data , x = tips_data['total_bill'], y = tips_data['tip'])
```

Scatter Plot: Total Bill vs. Tip Amount Categorized by Gender

```python
sns.scatterplot(data= tips_data, x = tips_data['total_bill'], y = tips_data['tip'], hue= tips_data['sex'])
```

Histogram: Age Distribution

```python
age = [45, 23, 58, 35, 30, 27, 65, 72, 18]
sns.histplot(age)
```

Histogram: Distribution of Total Bill Amounts

```python
sns.histplot(tips_data['total_bill'])
plt.show()
```

Histogram: Distribution of Tip Amounts

```python
sns.histplot(tips_data['tip'])
```

KDE Plot: Density Distribution of Total Bill Amounts"

```python
sns.kdeplot(tips_data['total_bill'])
```

Distribution Plot: Total Bill Amounts

```python
sns.distplot(tips_data['total_bill'])
```

Joint Plot: Relationship Between Total Bill and Tip Amounts

```python
sns.jointplot(data= tips_data , x = tips_data['total_bill'], y = tips_data['tip'])
```

Pair Plot: Relationships and Distributions in Tips Dataset

```python
sns.pairplot(data=tips_data)
```

Count Plot: Frequency of Meal Times

```python
sns.countplot(data= tips_data , x = tips_data['time'])
```

Count Plot: Meal Times Categorized by Gender

```python
sns.countplot(data= tips_data , x = tips_data['time'], hue= tips_data['sex'])
```

Bar Plot: Average Total Bill by Day

```python
sns.barplot(data= tips_data, x = tips_data['day'], y = tips_data['total_bill'])
```
