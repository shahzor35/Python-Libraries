```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
DF_1  = DataFrame(np.arange(36).reshape(6,6))
print(DF_1)
```

```python
DF_2 = DataFrame(np.arange(15).reshape(5,3))
print(DF_2)
```

```python
# Concation of DF_1 and DF_2
pd.concat([DF_1, DF_2], axis= 1)
```

```python
pd.concat([DF_1, DF_2])
```

```python
# Droping data in row wise
DF_1.drop([0,2]) # Only 0 and 2 row will be droped
```

```python
DF_1.drop([0,2], axis=1) # drop the column
```

```python
# Generating series data
series_col = Series(np.arange(6))
series_col.name = "Series column"
print(series_col)
```

```python
# Using JOIN
df_series_join = DataFrame.join(DF_1, series_col)
print(df_series_join)
```

```python
Result_of_combination = pd.concat([df_series_join, df_series_join])
print(Result_of_combination)
```

```python
Result_of_combination = pd.concat([df_series_join, df_series_join], ignore_index= True)
print(Result_of_combination)
```
