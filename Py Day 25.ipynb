 Extract the "mpg" (miles per gallon) column from the cars dataset into a variable named mpg. 

```python
mpg = cars.mpg
mpg
```

Create a detailed line plot to visualize the "Miles Per Gallon" (mpg) of cars in the dataset, with the following customizations:

Figure and Axes: Use add_axes to precisely control the plot's position and size.

Data: Plot the mpg values on the graph.

X-Ticks: Set custom tick labels for the x-axis using car names (cars.Cars_names), rotated 90 degrees for readability.

Titles and Labels: Add a descriptive title, x-axis label ("car names"), and y-axis label ("miles/gal").

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.Cars_names, rotation = 90, fontsize = 'medium')
ax.set_title('Miles Per Gallon of Cars in mtcars')
ax.set_xlabel('cars names')
ax.set_ylabel('miles/gal')
plt.show()
```
