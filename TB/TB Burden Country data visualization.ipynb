TB Burden Country data visualization 

```python
# Importing Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Dataset
TB_data = pd.read_csv("C:/Users/SHAHZOR AHMED/Downloads/TB_Burden_Country.csv")
TB_data
```

```python
# What is the total number of countries in the dataset?
TB_data['Country or territory name'].nunique()
```

```python
# What are the minimum, maximum, and average TB prevalence rates per 100,000 population globally?
# check if any null values
TB_data['Estimated prevalence of TB (all forms) per 100 000 population'].isnull().values.any()
```

```python
print(f"Minimum TB prevalence rates per 100,000 population globally: {TB_data['Estimated prevalence of TB (all forms) per 100 000 population'].dropna().min()}")
print(f"Maximum TB prevalence rates per 100,000 population globally: {TB_data['Estimated prevalence of TB (all forms) per 100 000 population'].dropna().max()}")
print(f"Average TB prevalence rates per 100,000 population globally: {TB_data['Estimated prevalence of TB (all forms) per 100 000 population'].dropna().mean()}")
```

```python
# How has the prevalence of TB per 100,000 population changed globally over the years?
plt.figure(figsize=(10, 6))
sns.lineplot(data = TB_data, x = 'Year', y = 'Estimated prevalence of TB (all forms) per 100 000 population')
plt.title("Global Trend of TB Prevalence Over the Years")
plt.show()
```

```python
# Listing Unique Country or Territory Names in the Dataset
TB_data['Country or territory name'].unique()
```

```python
# Select a specific country and visualize its TB prevalence trend over time.
India_TB_data = TB_data[TB_data['Country or territory name'] == 'India']
plt.figure(figsize=(10,6))
sns.lineplot(data= India_TB_data , x = 'Year', y = 'Estimated prevalence of TB (all forms) per 100 000 population')
plt.title("TB prevalence trend over time in India")
plt.grid(True)
plt.show()
```

```python
# Compare the average TB prevalence rates across different regions (e.g., EMR, AFR, etc.).
average_tb_by_region = TB_data.groupby('Region')['Estimated prevalence of TB (all forms) per 100 000 population'].mean()
Average_tb_by_region = pd.DataFrame(average_tb_by_region)
Average_tb_by_region
```

```python
# Bar Plot of Average TB Prevalence Rates Across Different Regions
plt.figure(figsize=(10,6))
sns.barplot(data= Average_tb_by_region, x = 'Region', y = 'Estimated prevalence of TB (all forms) per 100 000 population')
plt.title("Average TB prevalence rates across different regions")
plt.show()
```

```python
# Plot the relationship between population size and TB prevalence for different Region in a specific year.
specific_year = 1990
data_of_specific_year = TB_data[TB_data['Year'] == specific_year]
plt.figure(figsize=(12,6))
sns.scatterplot(data= data_of_specific_year , x = 'Estimated total population number', y = 'Estimated prevalence of TB (all forms) per 100 000 population', hue= 'Region')
plt.xscale('log')
plt.title(f"Relationship between Population Size and TB Prevalence ({specific_year})")
```

```python
# Use line plots to compare trends in TB prevalence between regions.
plt.figure(figsize=(10,6))
sns.lineplot(data= TB_data, x = 'Year', y = 'Estimated prevalence of TB (all forms) per 100 000 population', hue= 'Region')
plt.title("Comparing trends in TB prevalence between regions")
plt.show()
```

```python
# Use line plots to compare trends in TB prevalence between two or more regions.
specific_regions = ['EMR', 'SEA']
filtered_data_on_specific_regions = TB_data[TB_data['Region'].isin(specific_regions)]
pd.DataFrame(filtered_data_on_specific_regions)

plt.figure(figsize=(10,6))
sns.lineplot(data= filtered_data_on_specific_regions , x = 'Year', y = 'Estimated prevalence of TB (all forms) per 100 000 population', hue= 'Region')
plt.title(f"Trends in TB Prevalence Across {specific_regions}")
plt.show()
```
